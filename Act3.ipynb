```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Cargar datos de ejemplo
df = pd.DataFrame({
    'X': np.random.rand(100),
    'Y': np.random.rand(100),
    'Clase': np.random.choice([0, 1], 100)
})

# Separar datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(df[['X', 'Y']], df['Clase'], test_size=0.2)

# Modelo de Árbol de Decisión
arbol = DecisionTreeClassifier()
arbol.fit(X_train, y_train)

# Evaluación del modelo
print("Árbol de Decisión Score:", arbol.score(X_test, y_test))

# Visualización de los datos con clasificación
plt.scatter(df['X'], df['Y'], c=df['Clase'], cmap='coolwarm', alpha=0.6)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Clasificación con Árbol de Decisión')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Cargar datos de ejemplo
df = pd.DataFrame({
    'X': np.random.rand(100),
    'Y': np.random.rand(100),
    'Clase': np.random.choice([0, 1], 100)
})

# Separar datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(df[['X', 'Y']], df['Clase'], test_size=0.2)

# Modelo de Regresión Logística
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# Evaluación del modelo
print("Regresión Logística Score:", log_reg.score(X_test, y_test))

# Visualización de la frontera de decisión
plt.scatter(df['X'], df['Y'], c=df['Clase'], cmap='coolwarm', alpha=0.6)
x_values = np.linspace(df['X'].min(), df['X'].max(), 100)
y_values = -(log_reg.coef_[0][0] * x_values + log_reg.intercept_[0]) / log_reg.coef_[0][1]
plt.plot(x_values, y_values, color='red', linewidth=2, label='Frontera de Decisión')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión Logística')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Cargar datos de ejemplo
df = pd.DataFrame({
    'X': np.random.rand(100),
    'Y': np.random.rand(100),
})

# Modelo de Regresión Lineal
lin_reg = LinearRegression()
lin_reg.fit(df[['X']], df['Y'])

# Evaluación del modelo
print("Coeficientes Regresión Lineal:", lin_reg.coef_)

# Gráfica
plt.scatter(df['X'], df['Y'], color='blue', label='Datos')
plt.plot(df[['X']], lin_reg.predict(df[['X']]), color='red', linewidth=2, label='Línea de Regresión')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión Lineal')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Datos de ejemplo
ts_data = np.cumsum(np.random.randn(100))

# Modelo ARIMA
ts_model = ARIMA(ts_data, order=(2,1,2))
ts_result = ts_model.fit()

# Evaluación del modelo
print("Resumen ARIMA:\n", ts_result.summary())

plt.plot(ts_data, label='Datos Originales')
plt.plot(ts_result.fittedvalues, label='Predicción ARIMA', linestyle='dashed')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.title('Predicción con ARIMA')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Cargar datos de ejemplo
df = pd.DataFrame({
    'X': np.random.rand(100),
    'Y': np.random.rand(100),
})

# Modelo K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df[['X', 'Y']])
df['Cluster'] = kmeans.labels_

# Evaluación del modelo
print("Centroides K-Means:\n", kmeans.cluster_centers_)

# Visualización de Clustering
plt.scatter(df['X'], df['Y'], c=df['Cluster'], cmap='viridis', alpha=0.6)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroides')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Clustering con K-Means')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Datos de ejemplo
data = pd.DataFrame(np.random.randint(0, 2, size=(100, 5)), columns=['A', 'B', 'C', 'D', 'E'])

# Reglas de asociación 
freq_items = apriori(data, min_support=0.2, use_colnames=True)
rules = association_rules(freq_items, metric='confidence', min_threshold=0.5)

# Evaluación de las reglas
print("Reglas de Asociación:\n", rules)
```
